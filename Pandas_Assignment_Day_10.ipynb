## Applying functions to columns and creating new columns

We need to create another column in d3, which contains  a boolean value for each city to indicate whether it's a union territory or not.
- HINT: Chandigarh, Pondicherry and Delhi are only 3 union territories here.


```python
def is_ut(x):
    uts = ['Chandigarh', 'Pondicherry', 'Delhi']
    return x in uts

d3['is_ut'] = d3['city'].apply(is_ut)
print(d3)
```

The temperatures mentioned in ‘Temperature’ column are mentioned in Celsius, we need another column which contains the same in Fahrenheit.

HINT -
- Define a function c_to_f which takes input temp in celsius and returns a value with temperature in Fahrenheit.
- To check: c_to_f(10) should return 50.


```python
def c_to_f(c):
    return (c * 9/5) + 32
```

```python
print(c_to_f(10))
```

```python
d3['temp_farenhiet'] = d3['Temperature'].apply(c_to_f)
display(d3)
```

## Groupby

```python
# Create a dataframe using dictionary of your choice

# data =


```

```python
# Create a dataframe using dictionary of your choice
data = {
    'Category': ['Electronics', 'Furniture', 'Electronics', 'Furniture', 'Electronics', 'Appliances', 'Appliances'],
    'Sub-Category': ['Phones', 'Chairs', 'Laptops', 'Tables', 'Phones', 'Ovens', 'Microwaves'],
    'Sales': [1200, 800, 2500, 1500, 1100, 400, 300],
    'Quantity': [5, 2, 3, 4, 6, 1, 2]
}

df_group = pd.DataFrame(data)
display(df_group)
```

```python
# Use Groupby of single column with aggregate sum()


```

```python
# Use Groupby of single column with aggregate sum()
grouped_sum = df_group.groupby('Category')['Sales'].sum()
print(grouped_sum)
```

```python
# Use Groupby of single column with aggregate count()

```

```python
# Use Groupby of single column with aggregate count()
grouped_count = df_group.groupby('Category')['Quantity'].count()
print(grouped_count)
```

```python
grouped_min_max = df_group.groupby('Category')['Sales'].agg(['min', 'max'])
display(grouped_min_max)
```

```python
# Use Groupby of single column with aggregate min() and max()
grouped_min_max = df_group.groupby('Category')['Sales'].agg(['min', 'max'])
print(grouped_min_max)
```

```python
grouped_mean_multi = df_group.groupby(['Category', 'Sub-Category'])['Sales'].mean()
display(grouped_mean_multi)
```

```python
# Use Groupby of any 2 columns with aggregate mean()
grouped_mean_multi = df_group.groupby(['Category', 'Sub-Category'])['Sales'].mean()
print(grouped_mean_multi)
```

```python
grouped_min_max_multi = df_group.groupby(['Category', 'Sub-Category'])['Sales'].agg(['min', 'max'])
display(grouped_min_max_multi)
```

```python
# Use Groupby of any 2 columns with aggregate min() and max()
grouped_min_max_multi = df_group.groupby(['Category', 'Sub-Category'])['Sales'].agg(['min', 'max'])
print(grouped_min_max_multi)
```
