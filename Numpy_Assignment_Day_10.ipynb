## Numpy copy vs view

```python
x1 = np.array([1, 2, 3, 4, 5])
```

```python
x2 = x1
```

```python
print("x1:", x1)
print("x2:", x2)
```

Ok now you have seen that both of them are same

```python
x2[0] = 10
```

```python
print("x1 after changing x2:", x1)
print("x2 after changing x2:", x2)
```

```python
print("x1 and x2 share memory:", x1.base is x2.base)
```

Hey It's True they both share memory

Shall we try **view()** function also likwise.

```python
x3 = x1.view()
```

```python
print("x1 and x3 share memory:", x1.base is x3.base)
```

Woh! simple assignment is similar to view. That means
The view does not own the data and any changes made to the view will affect the original array, and any changes made to the original array will affect the view.

Don't agree? ok lets change x3 and see if original array i.e. x1 also changes

```python
x3[0] = 100
```

```python
print("x1 after changing x3:", x1)
print("x3 after changing x3:", x3)
```

Now its proved.

Lets see how **Copy()** function works

```python
x4 = x1.copy()
```

```python
x4[-1] = 900
```

```python
print("x1 after changing x4:", x1)
print("x4 after changing x4:", x4)
```

Hey! such an intresting output. You noticed buddy! your original array didn't get changed on change of its copy ie. x4.

Still not convinced? Ok lets see if they both share memory or not

```python
print("x1 and x4 share memory:", x1.base is x4.base)
```

**hstack vs vstack function**


Stacking is same as concatenation, the only difference is that stacking is done along a new axis.

NumPy provides a helper function:

1. hstack() to stack along rows.
2. vstack()  to stack along columns

```python
stacked_h = np.hstack((x1, x4))
print(stacked_h)
```

```python
stacked_v = np.vstack((x1, x4))
print(stacked_v)
```

We hope now you saw the difference between them.

Fun fact! you can even use concatenate() function to join 2 arrays along with the axis. If axis is not explicitly passed, it is taken as 0 ie. along column

Lets try this function as well

```python
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])

concatenated_rows = np.concatenate((arr1, arr2), axis=0)
print("Concatenated along rows:\n", concatenated_rows)
```

```python
concatenated_cols = np.concatenate((arr1, arr2), axis=1)
print("Concatenated along columns:\n", concatenated_cols)
```
